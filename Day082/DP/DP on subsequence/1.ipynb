```python
import matplotlib.pyplot as plt
import pandas as pd

# Data
data = {
    "Date": [
        "31-03-2026",
        "01-04-2026",
        "02-04-2026",
        "03-04-2026",
        "04-04-2026",
        "05-04-2026"
    ],
    "Axit": [142, 144, 149, 157, 158, 161],
    "Shikhar": [95, 98, 98, 99, 99, 103],
    "Arpit": [125, 125, 128, 131, 141, 145],
    "Yash": [123, 123, 126, 126, 126, 126],
    "Divyanshu": [65, 69, 70, 71, 71, 73]
}

df = pd.DataFrame(data)
df["Date"] = pd.to_datetime(df["Date"], format="%d-%m-%Y")

# Plot
plt.figure(figsize=(10, 5))

for name in df.columns[1:]:
    plt.plot(df["Date"], df[name], linewidth=2)

# Minimal styling
plt.title("Progress Tracker", fontsize=14)
plt.xticks(rotation=30)
plt.grid(alpha=0.3)

# Remove top & right borders (clean look)
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Add labels at last point instead of legend
for name in df.columns[1:]:
    plt.text(df["Date"].iloc[-1], df[name].iloc[-1], f" {name}", fontsize=10)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.interpolate import make_interp_spline

# Data
data = {
    "Date": [
        "20-01-2026",
        "31-03-2026",
        "01-04-2026",
        "02-04-2026",
        "03-04-2026",
        "04-04-2026",
        "05-04-2026"
    ],
    "Axit": [32, 142, 144, 149, 157, 158, 161],
    "Shikhar": [9, 95, 98, 98, 99, 99, 103],
    "Arpit": [0, 125, 125, 128, 131, 141, 145],
    "Yash": [0, 123, 123, 126, 126, 126, 126],
    "Divyanshu": [0, 65, 69, 70, 71, 71, 73]
}

df = pd.DataFrame(data)
df["Date"] = pd.to_datetime(df["Date"], format="%d-%m-%Y")

# Convert dates to numbers for smoothing
x = np.arange(len(df))

plt.figure(figsize=(10, 5))

# Smooth plot
for name in df.columns[1:]:
    y = df[name].values

    # Create smooth curve
    x_smooth = np.linspace(x.min(), x.max(), 300)
    spline = make_interp_spline(x, y, k=2)  # k=2 for smooth but stable
    y_smooth = spline(x_smooth)

    plt.plot(x_smooth, y_smooth, linewidth=2)

    # Label at end
    plt.text(x_smooth[-1], y_smooth[-1], f" {name}", fontsize=10)

# Styling (Apple-like clean)
plt.title("Progress Tracker", fontsize=14)
plt.xticks(x, df["Date"].dt.strftime("%d %b"), rotation=30)
plt.grid(alpha=0.2)

# Remove borders
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from matplotlib.animation import FuncAnimation

# Data
data = {
    "Date": ["31-03-2026","01-04-2026","02-04-2026","03-04-2026","04-04-2026","05-04-2026"],
    "Axit": [142,144,149,157,158,161],
    "Shikhar": [95,98,98,99,99,103],
    "Arpit": [125,125,128,131,141,145],
    "Yash": [123,123,126,126,126,126],
    "Divyanshu": [65,69,70,71,71,73]
}

df = pd.DataFrame(data)

fig, ax = plt.subplots(figsize=(10,5))
x = np.arange(len(df))

colors = {
    "Axit": "cyan",
    "Shikhar": "red",
    "Arpit": "yellow",
    "Yash": "green",
    "Divyanshu": "purple"
}

def update(frame):
    ax.clear()

    for name in df.columns[1:]:
        y = df[name][:frame+1]
        x_vals = x[:frame+1]

        # glow effect
        for i in range(1,5):
            ax.plot(x_vals, y, linewidth=2+i*2, alpha=0.05)

        # main line
        ax.plot(x_vals, y, linewidth=2)

        # label
        ax.text(x_vals[-1], y.iloc[-1], name)

    ax.set_title("DSA Progress Race 🚀")
    ax.set_xticks(x)
    ax.set_xticklabels(df["Date"], rotation=30)
    ax.grid(alpha=0.3)

ani = FuncAnimation(fig, update, frames=len(df), interval=800, repeat=False)

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Data
data = {
    "Date": ["31-03-2026","01-04-2026","02-04-2026","03-04-2026","04-04-2026","05-04-2026"],
    "Axit": [142,144,149,157,158,161],
    "Shikhar": [95,98,98,99,99,103],
    "Arpit": [125,125,128,131,141,145],
    "Yash": [123,123,126,126,126,126],
    "Divyanshu": [65,69,70,71,71,73]
}

df = pd.DataFrame(data)

# Convert Date
df["Date"] = pd.to_datetime(df["Date"], format="%d-%m-%Y")

# Calculate daily solved (difference)
daily = df.copy()
for col in df.columns[1:]:
    daily[col] = df[col].diff()

# First day will be NaN → replace with 0
daily.fillna(0, inplace=True)

# Plot
plt.figure(figsize=(10,5))

for name in daily.columns[1:]:
    plt.plot(daily["Date"], daily[name], marker='o', linewidth=2, label=name)

plt.title("Daily Questions Solved 📊")
plt.xlabel("Date")
plt.ylabel("Questions Solved")
plt.xticks(rotation=30)
plt.grid(alpha=0.3)
plt.legend()

plt.tight_layout()
plt.show()
```
