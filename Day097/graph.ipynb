```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline

# Dark theme
plt.style.use("dark_background")

# Data (excluding last row)
dates = [
    "31-03","01-04","02-04","03-04","04-04","05-04","06-04","07-04",
    "08-04","09-04","10-04","11-04","12-04","13-04","14-04","15-04",
    "16-04","17-04","18-04","19-04"
]

x = np.arange(len(dates))

data = {
    "Axit": [142,144,149,157,158,161,166,167,168,169,171,176,183,187,193,197,198,199,200,201],
    "Shikhar": [95,98,98,99,99,103,107,108,109,110,111,112,113,115,116,117,117,118,119,121],
    "Arpit": [125,125,128,131,141,145,147,148,149,150,153,154,156,158,163,166,172,175,177,179],
    "Yash": [123,123,126,126,126,126,126,126,126,126,126,126,130,130,133,134,135,135,135,135],
    "Divyanshu": [65,69,70,71,71,73,74,78,79,81,83,87,90,91,91,91,91,91,92,92]
}

colors = {
    "Axit": "#00FFCC",       # highlight color
    "Shikhar": "#FF6B6B",
    "Arpit": "#FFD93D",
    "Yash": "#6C5CE7",
    "Divyanshu": "#00B894"
}

plt.figure(figsize=(14,7))

for name, y in data.items():
    y = np.array(y)

    # Smooth curve
    x_smooth = np.linspace(x.min(), x.max(), 300)
    spline = make_interp_spline(x, y, k=3)
    y_smooth = spline(x_smooth)

    # Glow effect (draw multiple transparent lines)
    for i in range(1, 6):
        plt.plot(x_smooth, y_smooth, color=colors[name], alpha=0.05*i, linewidth=2+i)

    # Main line
    lw = 3 if name == "Axit" else 2
    plt.plot(x_smooth, y_smooth, color=colors[name], linewidth=lw)

    # Label at end
    plt.text(x_smooth[-1] + 0.3, y_smooth[-1], name,
             color=colors[name],
             fontsize=11,
             weight='bold' if name == "Axit" else 'normal',
             va='center')

# Grid styling
plt.grid(alpha=0.15)

# Axis styling
plt.xticks(x, dates, rotation=45)
plt.xlabel("Date", fontsize=12, alpha=0.8)
plt.ylabel("Questions Solved", fontsize=12, alpha=0.8)

# Title
plt.title("📈 Daily Progress Tracker", fontsize=18, weight='bold', pad=20)

# Remove top/right borders
for spine in ["top", "right"]:
    plt.gca().spines[spine].set_visible(False)

plt.tight_layout()
plt.show()
```
