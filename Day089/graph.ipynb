```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import make_interp_spline

# Data from your sheet
dates = [
    "31-03", "01-04", "02-04", "03-04", "04-04",
    "05-04", "06-04", "07-04", "08-04", "09-04",
    "10-04", "11-04"
]

# Total questions (Axit column)
questions = [142, 144, 149, 157, 158, 161, 166, 167, 168, 169, 171, 176]

# Convert x-axis to numeric
x = np.arange(len(dates))

# -------------------------------
# 1. Smooth Graph: Date vs Total Questions
# -------------------------------
x_smooth = np.linspace(x.min(), x.max(), 300)
spl = make_interp_spline(x, questions, k=3)
y_smooth = spl(x_smooth)

plt.figure()
plt.plot(x_smooth, y_smooth)
plt.xticks(x, dates)
plt.xlabel("Date")
plt.ylabel("Total Questions")
plt.title("Progress Over Time")
plt.grid()
plt.show()
```

```python

# -------------------------------
# 2. Per Day Questions Graph
# -------------------------------
daily_questions = [questions[i] - questions[i-1] for i in range(1, len(questions))]
dates_daily = dates[1:]
x2 = np.arange(len(daily_questions))

# Smooth curve
x2_smooth = np.linspace(x2.min(), x2.max(), 300)
spl2 = make_interp_spline(x2, daily_questions, k=3)
y2_smooth = spl2(x2_smooth)

plt.figure()
plt.plot(x2_smooth, y2_smooth)
plt.xticks(x2, dates_daily)
plt.xlabel("Date")
plt.ylabel("Questions per Day")
plt.title("Daily Performance")
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import make_interp_spline
from matplotlib.animation import FuncAnimation

dates = [
    "31-03", "01-04", "02-04", "03-04", "04-04",
    "05-04", "06-04", "07-04", "08-04", "09-04",
    "10-04", "11-04"
]

data = {
    "Axit": [142, 144, 149, 157, 158, 161, 166, 167, 168, 169, 171, 176],
    "Shikhar": [95, 98, 98, 99, 99, 103, 107, 108, 109, 110, 111, 112],
    "Arpit": [125, 125, 128, 131, 141, 145, 147, 148, 149, 150, 153, 154],
    "Yash": [123, 123, 126, 126, 126, 126, 126, 126, 126, 126, 126, 126],
    "Divyanshu": [65, 69, 70, 71, 71, 73, 74, 78, 79, 81, 83, 87]
}

x = np.arange(len(dates))

fig, ax = plt.subplots(figsize=(10, 6))

def update(frame):
    ax.clear()
    
    for name, values in data.items():
        x_part = x[:frame+1]
        y_part = values[:frame+1]

        if len(x_part) > 3:
            x_smooth = np.linspace(x_part.min(), x_part.max(), 100)
            spline = make_interp_spline(x_part, y_part, k=2)
            y_smooth = spline(x_smooth)
            ax.plot(x_smooth, y_smooth)
        else:
            ax.plot(x_part, y_part)

        ax.text(x_part[-1], y_part[-1], f" {name}")

    ax.set_xticks(x)
    ax.set_xticklabels(dates, rotation=30)
    ax.set_title(f"Progress till {dates[frame]}")
    ax.set_xlabel("Date")
    ax.set_ylabel("Questions Solved")
    ax.grid()

ani = FuncAnimation(fig, update, frames=len(dates), interval=800)
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

dates = [
    "31-03", "01-04", "02-04", "03-04", "04-04",
    "05-04", "06-04", "07-04", "08-04", "09-04",
    "10-04", "11-04"
]

data = {
    "Axit": [142, 144, 149, 157, 158, 161, 166, 167, 168, 169, 171, 176],
    "Shikhar": [95, 98, 98, 99, 99, 103, 107, 108, 109, 110, 111, 112],
    "Arpit": [125, 125, 128, 131, 141, 145, 147, 148, 149, 150, 153, 154],
    "Yash": [123, 123, 126, 126, 126, 126, 126, 126, 126, 126, 126, 126],
    "Divyanshu": [65, 69, 70, 71, 71, 73, 74, 78, 79, 81, 83, 87]
}

names = list(data.keys())

fig, ax = plt.subplots(figsize=(10, 6))

def update(frame):
    ax.clear()

    # Get values for this day
    current = {name: data[name][frame] for name in names}

    # Sort by score
    sorted_data = sorted(current.items(), key=lambda x: x[1])

    names_sorted = [x[0] for x in sorted_data]
    values_sorted = [x[1] for x in sorted_data]

    ax.barh(names_sorted, values_sorted)

    ax.set_title(f"🏆 Ranking on {dates[frame]}")
    ax.set_xlabel("Total Questions")

    for i, v in enumerate(values_sorted):
        ax.text(v + 1, i, str(v))

ani = FuncAnimation(fig, update, frames=len(dates), interval=800)
plt.show()
```
